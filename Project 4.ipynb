```python
def scrape_jobspage(search):
    
    allrows = []

    for jobtitle in search:
        
        row = []

        for page in range(0,100,10):

            url = 'https://www.indeed.com.sg/jobs?q=' + jobtitle.replace(' ','+')+ '&l=Singapore&start=' + str(page)
            response = requests.get(url)
            html = response.text
            soup = BeautifulSoup(html, 'lxml')

            get_title = soup.find("h2", {"class":"jobtitle"}).text
            row.append(get_title)

            get_company = soup.find("span", {"class":"company"}).text
            row.append(get_company)

            try:
                get_location = soup.find("span", {"class":"location"}).text
            except:
                get_location = np.nan

            row.append(get_location)

            try:
                get_summary = soup.find("span", {"id":"job_summary"}).text
            except:
                get_summary = np.nan

            row.append(get_summary)

            get_extra = soup.findAll("span", {"class":"no-wrap"})

            try:
                for info in get_extra:
                    for char in info.text.encode('ascii','ignore'):
                        if char != '$':
                            job_type = info.text.encode('ascii','ignore')

                        if char == '$':
                            salary_range = info.text.encode('ascii','ignore')
                            salary = re.findall(r'\d+',salary_range.replace(',',''))
                            salary = np.mean([float(s) for s in salary])

                            break
            except:
                job_type = np.nan
                salary = np.nan


            row.append(job_type)
            row.append(salary)
            
            allrows.append(row)
            
            print "running page:" + str(page)
            
            time.sleep(3)

    return allrows
    
```

```python
scraped = scrape_jobspage(['data scientist'])
```
